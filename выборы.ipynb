1. Просмотр датасета

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.ticker as plticker
import csv

data_path = f'/content/drive/MyDrive/Colab Notebooks/130921.csv'
data = pd.read_csv(data_path, sep=';')
data.info()
```

```python
data.head(len(data))
```

2. Оставляем только необходимые данные

```python
data = data[['Name','Region_subject','Voters_in_list', 'Number_of_valid_bulletins', 'Votes_for_Aitakova', 
             'Votes_for_Kozlov', 'Votes_for_Makhonin', 'Votes_for_Postnikov','Leading_candidate']]
data.to_csv("130921.csv", index=False)
#data.head(len(data))
data.info()
```

3. Добавляем колонку данных, необходимых для анализа

```python
lst = ["Итого", "Итого"]    
summ = data.sum().values
for i in range(2,len(summ)-1):
  lst.append(summ[i])
lst.append("Махонин")
data.loc[len(data)] = lst
data.head(len(data))
```

```python
data['Percent_voter_turnout'] = round(data['Number_of_valid_bulletins'] / data['Voters_in_list'] * 100, 2)   # получим явку в процентах для каждого ТИК
data.head(len(data))
```

```python
data['Votes_for_Aitakova'] = round(data['Votes_for_Aitakova'] / data['Number_of_valid_bulletins'] * 100, 2)
data['Votes_for_Kozlov'] = round(data['Votes_for_Kozlov'] / data['Number_of_valid_bulletins'] * 100, 2)
data['Votes_for_Makhonin'] = round(data['Votes_for_Makhonin'] / data['Number_of_valid_bulletins'] * 100, 2)
data['Votes_for_Postnikov'] = round(data['Votes_for_Postnikov'] / data['Number_of_valid_bulletins'] * 100, 2)
data.head(len(data))
```

7. Агрегируеми сортируем

```python
data.groupby('Region_subject').agg('mean').sort_values('Percent_voter_turnout')[['Percent_voter_turnout']]
```

8. Диаграмма рассеяния

```python
fig, ax = plt.subplots(figsize=(10, 10))
ax.scatter(x = data['Percent_voter_turnout'], y = data['Votes_for_Makhonin'])
plt.xlabel("Percent_voter_turnout")
plt.ylabel("Votes_for_Makhonin")
#stats2.plot.scatter("Percent_voter_turnout", "Votes_for_Makhonin", figsize=(15, 15))
```

10.Гистограмма

```python
fig,ax=plt.subplots()
plt.hist(data['Percent_voter_turnout'],bins=np.arange(0,100,0.5), align='mid',density = True)
plt.grid()
plt.xlabel('Явка, %')
plt.ylabel('Плотность распределения')
intervals = 5
loc = plticker.MultipleLocator(base=intervals)
ax.xaxis.set_major_locator(loc)
ax.grid(which='major', axis='both', linestyle='-')
```
